# Crime Rates

Here, we are determining which neighborhood is the safest neighborhood to raise a child. 

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import geopandas
```

We import the data and display the first ten rows.

```python
data = pd.read_csv("PittsburghPoliceArrestData.csv")

# look at the first ten rows of the data
data.head(10)
```

Let's get a sum for each neighborhood of the amount of crimes that occured in them. Each row is one crime committed in a particular neighborhood. Therefore, if we count the number of rows that have a certain incident neighborhood, then we get the total crimes committed in that neighborhood.

```python
data['INCIDENTNEIGHBORHOOD'].value_counts()
crime = pd.DataFrame() 
crime["incidents"] = data['INCIDENTNEIGHBORHOOD'].value_counts()
crime = crime.reset_index()
#crime["incidents"] = data['INCIDENTNEIGHBORHOOD'].value_counts()
crime
```

Let's plot this data!

```python
data['INCIDENTNEIGHBORHOOD'].value_counts().plot()
```

Ew! The axis is hard to read! Let's make a bar graph instead!

```python
plt.style.use('classic')
data['INCIDENTNEIGHBORHOOD'].value_counts().plot.bar(title='Number of Crimes in a Neighborhood', figsize=(20,20))
```

This is a headache inducing bar graph, but it is legable. The neighborhoods with the most crime recorded was the Central Business District, South Side Flats Carrick, East Allegheny and Homewood South.

```python
neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods_.shp")
neighborhoods.plot() # plot the geopandas dataframe
```

```python
# do the merge
#num_crimes=data.groupby("INCIDENTNEIGHBORHOOD").size()
#num_crimes.sort_values(ascending=False)
#crime_map = neighborhoods.merge(num_crimes.to_frame(), how='left', left_on='hood', right_on="INCIDENTNEIGHBORHOOD")
hood = neighborhoods.merge(crime, how='left', left_on='hood', right_on="index")
# look at the head to confirm it merged correctly
hood = hood.dropna()
hood[['hood',"incidents",'geometry']].head()
#steps_map = neighborhoods.merge(num_steps, how='left', left_on='hood', right_on='neighborhood')
```
